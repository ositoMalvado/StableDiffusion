````python
import zipfile
import os
from IPython.display import HTML, display
def mostrar_en_div_estilizado(contenido):
    div_style = """
        background-color: #007BFF;
        color: #FFFFFF;
        padding: 15px;
        border-radius: 5px;
        font-weight: bold;
        font-family: Arial, sans-serif;
    """
    div = f'<div class="alert alert-primary" style="{div_style}">{contenido}</div>'
    display(HTML(div))
def comprimir_directorio_en_zip(directorio, archivo_zip):
    with zipfile.ZipFile(archivo_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for carpeta_actual, subcarpetas, archivos in os.walk(directorio):
            for archivo in archivos:
                ruta_completa = os.path.join(carpeta_actual, archivo)
                ruta_relativa = os.path.relpath(ruta_completa, directorio)
                zipf.write(ruta_completa, ruta_relativa)
comprimir_directorio_en_zip('/home/studio-lab-user/stable-diffusion-webui/outputs', '/home/studio-lab-user/imagenes.zip')
mostrar_en_div_estilizado(f'Se guardaron las im√°genes en "/home/studio-lab-user/imagenes.zip"')
````
